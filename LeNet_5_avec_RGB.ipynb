```python
import numpy as np
import pandas as pd
import kagglehub
import os
import sys
import cv2
import csv
import random
import psutil
import logging
import matplotlib.pyplot as plt
from tqdm import trange, tqdm
from joblib import Parallel, delayed
from sklearn.model_selection import KFold, train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder



logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s', handlers=[logging.FileHandler('pipeline.log', mode='w'), logging.StreamHandler(sys.stdout)], force=True)

# Fonctions utilitaires pour le chargement et le prétraitement des images
def load_and_preprocess_image(image_path, data_dir, target_size=(32, 32)):
    full_path = os.path.join(data_dir, image_path)
    img = cv2.imread(full_path, cv2.IMREAD_GRAYSCALE)  #cahrgement des images niveaux de gris
    if img is None:
        raise ValueError(f"Échec du chargement : {full_path}")
    img = cv2.resize(img, target_size, interpolation=cv2.INTER_AREA)  # Redimensionner à la taille cible
    img = img.astype(np.float32) / 255.0  # Normaliser les pixels entre 0 et 1
    return img.flatten()  # Aplatir l'image en un vecteur 1D

def process_image(idx, path, data_dir, target_size):
    try:
        img = load_and_preprocess_image(path, data_dir, target_size)
        return idx, img
    except Exception as e:
        logging.error(f"Échec du traitement de {path} : {e}")
        return idx, None

def parallel_load_images(image_paths, data_dir, target_size=(32, 32), n_jobs=32, chunk_size=512):
    image_paths = list(image_paths)
    n_samples = len(image_paths)
    all_images = []
    all_indices = []
    n_jobs = min(n_jobs, max(1, psutil.cpu_count(logical=False) - 1))  # Ajuster n_jobs dynamiquement

    for start in range(0, n_samples, chunk_size):
        end = min(start + chunk_size, n_samples)
        chunk_paths = image_paths[start:end]
        results = Parallel(n_jobs=n_jobs, backend='threading')(
            delayed(process_image)(idx + start, path, data_dir, target_size)
            for idx, path in enumerate(chunk_paths)
        )
        results.sort(key=lambda x: x[0])
        valid_images = [r[1] for r in results if r[1] is not None]
        valid_indices = [r[0] for r in results if r[1] is not None]
        all_images.extend(valid_images)
        all_indices.extend(valid_indices)
        logging.info(f"Processed chunk {start//chunk_size + 1}/{(n_samples + chunk_size - 1)//chunk_size}")

    X_array = np.array(all_images, dtype=np.float32)
    return X_array, np.array(all_indices)

# Augmentation des données
def augment_image(image, max_rotation=5, max_translation=3, seed=None):
    try:
        assert image.shape == (32, 32), f"Expected image shape (32, 32), got {image.shape}"
        rng = np.random.default_rng(seed)
        center = (16, 16)  # Centre de rotation
        # Rotation
        angle = rng.uniform(-max_rotation, max_rotation)
        M = cv2.getRotationMatrix2D(center, angle, 1.0)
        rotated = cv2.warpAffine(image, M, (32, 32), borderMode=cv2.BORDER_REPLICATE)
        tx = rng.uniform(-max_translation, max_translation)
        M = np.float32([[1, 0, tx], [0, 1, 0]])
        translated = cv2.warpAffine(rotated, M, (32, 32), borderMode=cv2.BORDER_REPLICATE)
        return translated.flatten()
    except AssertionError as e:
        logging.error(f"augment_image failed: {e}")
        raise
    except Exception as e:
        logging.exception(f"Unexpected error during augment_image: {e}")
        raise

def pre_augment_dataset(X, y, augmentations_per_image=2, n_jobs=32, chunk_size=512):
    X_np = X
    y_np = y
    n_samples = X_np.shape[0]
    augmented_X = []
    augmented_y = []
    base_seed = 1000
    n_jobs = min(n_jobs, max(1, psutil.cpu_count(logical=False) - 1))

    def augment_single(img, label, idx):
        rng = np.random.default_rng(base_seed + idx)
        results = [img]
        results.extend(
            augment_image(img.reshape(32, 32), seed=base_seed + idx * (i + 1))
            for i in range(augmentations_per_image)
        )
        return results, [label] * (augmentations_per_image + 1)

    for start in range(0, n_samples, chunk_size):
        end = min(start + chunk_size, n_samples)
        chunk_X = X_np[start:end]
        chunk_y = y_np[start:end]
        tasks = [(chunk_X[i], chunk_y[i], start + i) for i in range(len(chunk_X))]
        results = Parallel(n_jobs=n_jobs, backend='loky')(
            delayed(augment_single)(img, label, idx) for img, label, idx in tasks
        )
        for chunk_images, chunk_labels in results:
            augmented_X.extend(chunk_images)
            augmented_y.extend(chunk_labels)
        logging.info(f"Augmented chunk {start//chunk_size + 1}/{(n_samples + chunk_size - 1)//chunk_size}")

    return np.array(augmented_X), np.array(augmented_y)

# Recherche aléatoire d'hyperparamètres
def random_search(param_distributions, n_iter=5, seed=42):
    random.seed(seed)
    np.random.seed(seed)
    optimizers = param_distributions['optimizer']
    other_params = {k: v for k, v in param_distributions.items() if k != 'optimizer'}
    keys = sorted(other_params.keys())
    unique_combinations = []
    seen = set()
    per_optimizer = max(1, n_iter // len(optimizers))

    for opt in optimizers:
        count = 0
        attempts = 0
        max_attempts = per_optimizer * 10
        while count < per_optimizer and attempts < max_attempts:
            comb = {'optimizer': opt}
            for k in keys:
                v = other_params[k]
                if isinstance(v, tuple):
                    comb[k] = np.random.uniform(v[0], v[1])
                else:
                    comb[k] = random.choice(v)
            key = tuple((kk, round(vv, 8) if isinstance(vv, float) else vv) for kk, vv in sorted(comb.items()))
            if key not in seen:
                seen.add(key)
                unique_combinations.append(comb)
                count += 1
            attempts += 1

    while len(unique_combinations) < n_iter:
        comb = {'optimizer': random.choice(optimizers)}
        for k in keys:
            v = other_params[k]
            if isinstance(v, tuple):
                comb[k] = np.random.uniform(v[0], v[1])
            else:
                comb[k] = random.choice(v)
        key = tuple((kk, round(vv, 8) if isinstance(vv, float) else vv) for kk, vv in sorted(comb.items()))
        if key not in seen:
            seen.add(key)
            unique_combinations.append(comb)

    logging.info(f"Generated {len(unique_combinations)} hyperparameter combinations")
    return unique_combinations

# Réseau de neurones multiclasse
class MulticlassNeuralNetwork:
    def __init__(self, layer_sizes, learning_rate=0.001, l2_lambda=0.01, batch_size=128, optimizer='adam', momentum=0.9, beta1=0.9, beta2=0.999, epsilon=1e-8):
        self.layer_sizes = layer_sizes
        self.learning_rate = learning_rate
        self.l2_lambda = l2_lambda
        self.batch_size = batch_size
        self.optimizer = optimizer
        self.momentum = momentum
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.t = 0  # Compteur d'itérations pour Adam
        # Initialisation des poids avec la méthode He et des biais à zéro
        self.weights = [np.random.randn(layer_sizes[i], layer_sizes[i+1]) * np.sqrt(2/layer_sizes[i]) for i in range(len(layer_sizes)-1)]
        self.biases = [np.zeros((1, layer_sizes[i+1])) for i in range(len(layer_sizes)-1)]
        # Initialisation des moments pour SGD avec momentum et Adam
        self.m_weights = [np.zeros_like(w) for w in self.weights]
        self.m_biases = [np.zeros_like(b) for b in self.biases]
        self.v_weights = [np.zeros_like(w) for w in self.weights]  # Pour Adam
        self.v_biases = [np.zeros_like(b) for b in self.biases]   # Pour Adam

    def relu(self, x):
        return np.maximum(0, x)

    def relu_derivative(self, x):
        return np.where(x > 0, 1, 0)

    def softmax(self, x):
        exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
        return exp_x / np.sum(exp_x, axis=1, keepdims=True)

    def forward(self, X):
        self.z_values, self.activations = [], [X]
        for i in range(len(self.weights) - 1):
            Z = np.dot(self.activations[-1], self.weights[i]) + self.biases[i]
            self.z_values.append(Z)
            self.activations.append(self.relu(Z))
        Z = np.dot(self.activations[-1], self.weights[-1]) + self.biases[-1]
        self.z_values.append(Z)
        self.activations.append(self.softmax(Z))
        return self.activations[-1]

    def compute_loss(self, y_true, y_pred):
        m = y_true.shape[0]
        y_pred = np.clip(y_pred, 1e-15, 1 - 1e-15)
        cross_entropy = -np.sum(y_true * np.log(y_pred)) / m
        l2_term = (self.l2_lambda / (2 * m)) * sum(np.sum(np.square(W)) for W in self.weights)
        return cross_entropy + l2_term

    def compute_accuracy(self, y_true, y_pred):
        predictions = np.argmax(y_pred, axis=1)
        true_labels = np.argmax(y_true, axis=1)
        return np.mean(predictions == true_labels)

    def backward(self, X, y):
        m = X.shape[0]
        self.d_weights = [np.zeros_like(w) for w in self.weights]
        self.d_biases = [np.zeros_like(b) for b in self.biases]
        dZ = self.activations[-1] - y
        self.d_weights[-1] = (np.dot(self.activations[-2].T, dZ) + self.l2_lambda * self.weights[-1]) / m
        self.d_biases[-1] = np.sum(dZ, axis=0, keepdims=True) / m
        dA_prev = dZ
        for i in range(len(self.weights) - 2, -1, -1):
            dA = np.dot(dA_prev, self.weights[i + 1].T)
            dZ = dA * self.relu_derivative(self.z_values[i])
            self.d_weights[i] = (np.dot(self.activations[i].T, dZ) + self.l2_lambda * self.weights[i]) / m
            self.d_biases[i] = np.sum(dZ, axis=0, keepdims=True) / m
            dA_prev = dZ

    def update_parameters(self):
        if self.optimizer == 'sgd':
            for i in range(len(self.weights)):
                self.m_weights[i] = self.momentum * self.m_weights[i] - self.learning_rate * self.d_weights[i]
                self.m_biases[i] = self.momentum * self.m_biases[i] - self.learning_rate * self.d_biases[i]
                self.weights[i] += self.m_weights[i]
                self.biases[i] += self.m_biases[i]
        elif self.optimizer == 'adam':
            self.t += 1
            for i in range(len(self.weights)):
                self.m_weights[i] = self.beta1 * self.m_weights[i] + (1 - self.beta1) * self.d_weights[i]
                self.v_weights[i] = self.beta2 * self.v_weights[i] + (1 - self.beta2) * (self.d_weights[i] ** 2)
                self.m_biases[i] = self.beta1 * self.m_biases[i] + (1 - self.beta1) * self.d_biases[i]
                self.v_biases[i] = self.beta2 * self.v_biases[i] + (1 - self.beta2) * (self.d_biases[i] ** 2)
                m_w_hat = self.m_weights[i] / (1 - self.beta1 ** self.t)
                v_w_hat = self.v_weights[i] / (1 - self.beta2 ** self.t)
                m_b_hat = self.m_biases[i] / (1 - self.beta1 ** self.t)
                v_b_hat = self.v_biases[i] / (1 - self.beta2 ** self.t)
                self.weights[i] -= self.learning_rate * m_w_hat / (np.sqrt(v_w_hat) + self.epsilon)
                self.biases[i] -= self.learning_rate * m_b_hat / (np.sqrt(v_b_hat) + self.epsilon)

    def train(self, X, y, X_val, y_val, epochs=100):
        train_losses, train_accuracies, val_losses, val_accuracies = [], [], [], []
        for epoch in range(epochs):
            indices = np.random.permutation(X.shape[0])
            X_shuffled, y_shuffled = X[indices], y[indices]
            for i in range(0, X.shape[0], self.batch_size):
                X_batch = X_shuffled[i:i+self.batch_size]
                y_batch = y_shuffled[i:i+self.batch_size]
                outputs = self.forward(X_batch)
                self.backward(X_batch, y_batch)
                self.update_parameters()
            train_pred = self.forward(X)
            val_pred = self.forward(X_val)
            train_loss = self.compute_loss(y, train_pred)
            val_loss = self.compute_loss(y_val, val_pred)
            train_acc = self.compute_accuracy(y, train_pred)
            val_acc = self.compute_accuracy(y_val, val_pred)
            train_losses.append(train_loss)
            val_losses.append(val_loss)
            train_accuracies.append(train_acc)
            val_accuracies.append(val_acc)
            logging.info(f"Epoch {epoch+1}/{epochs}: Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}")
        # Affichage des courbes
        self.plot_metrics(train_losses, val_losses, train_accuracies, val_accuracies)
        return train_losses, val_losses, train_accuracies, val_accuracies

    def plot_metrics(self, train_losses, val_losses, train_accuracies, val_accuracies):
        epochs = range(1, len(train_losses) + 1)
        plt.figure(figsize=(12, 5))

        # Courbe des pertes
        plt.subplot(1, 2, 1)
        plt.plot(epochs, train_losses, label='Train Loss')
        plt.plot(epochs, val_losses, label='Validation Loss')
        plt.title('Training and Validation Loss')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True)

        # Courbe des précisions
        plt.subplot(1, 2, 2)
        plt.plot(epochs, train_accuracies, label='Train Accuracy')
        plt.plot(epochs, val_accuracies, label='Validation Accuracy')
        plt.title('Training and Validation Accuracy')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.grid(True)

        plt.tight_layout()
        plt.show()

# Validation croisée
def train_fold(fold, X, y_one_hot, train_idx, val_idx, layer_sizes, epochs, params):
    X_train, X_val = X[train_idx], X[val_idx]
    y_train_one_hot, y_val_one_hot = y_one_hot[train_idx], y_one_hot[val_idx]
    nn = MulticlassNeuralNetwork(layer_sizes, **params)
    train_losses, val_losses, train_accs, val_accs = nn.train(
        X_train, y_train_one_hot, X_val, y_val_one_hot, epochs=epochs
    )
    return {"fold": fold + 1, "final_val_accuracy": val_accs[-1]}

def k_fold_cross_validation(X, y_one_hot, layer_sizes, k=5, epochs=100, params=None, parallel=True):
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
    folds_data = []
    n_jobs = min(k, max(1, psutil.cpu_count(logical=False) - 1))  # Ajuster n_jobs dynamiquement
    tasks = [(fold, X, y_one_hot, train_idx, val_idx, layer_sizes, epochs, params)
             for fold, (train_idx, val_idx) in enumerate(kf.split(X))]

    if parallel and k > 1:
        results = Parallel(n_jobs=n_jobs, backend='loky', return_as='generator')(
            delayed(train_fold)(*task) for task in tasks
        )
        folds_data = list(tqdm(results, total=len(tasks), desc="Cross-validation folds", unit="fold"))
    else:
        for task in tqdm(tasks, desc="Cross-validation folds", unit="fold"):
            folds_data.append(train_fold(*task))

    return {"folds": folds_data, "final_val_accuracy": np.mean([f["final_val_accuracy"] for f in folds_data])}

# Réglage des hyperparamètres
def evaluate_params(params, X, y_one_hot, layer_sizes, k=5, epochs=100, trial=1, verbose=1, parallel=True):
    logging.info(f"Trial {trial}: Params: {params}")
    result = k_fold_cross_validation(X, y_one_hot, layer_sizes, k, epochs, params, parallel=parallel)
    final_val_acc = result["final_val_accuracy"]
    folds = result["folds"]

    if verbose:
        for fold_data in folds:
            fold = fold_data["fold"]
            fold_acc = fold_data["final_val_accuracy"]
            logging.info(f"Trial {trial}, Fold {fold}: Val acc: {fold_acc:.4f}")

    logging.info(f"Trial {trial}: Final val acc: {final_val_acc:.4f}")
    return final_val_acc, result

# Exécution principale
path = kagglehub.dataset_download("benaddym/amazigh-handwritten-character-database-amhcd")
target_dir = path
logging.info(f"Dataset path: {target_dir}")

latin_labels_txt = os.path.join(target_dir, 'labels/labels/33-common-latin-tifinagh.txt')
tifinagh_labels_txt = os.path.join(target_dir, 'labels/labels/sorted-33-common-tifinagh.txt')
images_root = os.path.join(target_dir, 'AMHCD_64/AMHCD_64')
output_csv = os.path.join('/content', 'rgb-labels-map.csv')

with open(latin_labels_txt, 'r', encoding='utf-8') as f:
    latin_labels = [line.strip() for line in f if line.strip()]
with open(tifinagh_labels_txt, 'r', encoding='utf-8') as f:
    tif_chars = [line.strip() for line in f if line.strip()]
if len(latin_labels) != len(tif_chars):
    logging.error(f"Error: {len(latin_labels)} Latin labels ≠ {len(tif_chars)} Tifinagh")
    sys.exit(1)
with open(output_csv, 'w', encoding='utf-8', newline='') as out:
    writer = csv.writer(out)
    for latin, tif in zip(latin_labels, tif_chars):
        folder = os.path.join(images_root, latin)
        if not os.path.isdir(folder):
            logging.warning(f"Skipping: {folder}")
            continue
        for fname in sorted(os.listdir(folder)):
            if not fname.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff')):
                continue
            rel_path = os.path.join(folder, fname)
            writer.writerow([rel_path, tif])
logging.info(f"Generated {output_csv} with {sum(1 for _ in open(output_csv))} entries.")

# Chargement des données
labels_df = pd.read_csv(output_csv, names=['image_path', 'label'])
labels_df = labels_df.iloc[:5000]
label_encoder = LabelEncoder()
labels_df['label_encoded'] = label_encoder.fit_transform(labels_df['label'])
num_classes = len(label_encoder.classes_)

logging.info("Loading dataset...")
X, valid_indices = parallel_load_images(labels_df['image_path'], target_dir, target_size=(32, 32))
labels_df = labels_df.iloc[valid_indices].reset_index(drop=True)
y = labels_df['label_encoded'].values

logging.info("Augmenting dataset...")
X_aug, y_aug = pre_augment_dataset(X, y, augmentations_per_image=1)

X_train, X_test, y_train, y_test = train_test_split(X_aug, y_aug, test_size=0.2, stratify=y_aug, random_state=42)
one_hot = OneHotEncoder(sparse_output=False)
y_train_one_hot = one_hot.fit_transform(y_train.reshape(-1, 1))
y_test_one_hot = one_hot.transform(y_test.reshape(-1, 1))

layer_sizes = [32 * 32, 64, 32, num_classes]
param_distributions = {
    'batch_size': [128, 256],
    'l2_lambda': [0.0001, 0.001],
    'learning_rate': [0.001, 0.0001],
    'optimizer': ['sgd', 'adam'],
}
n_iter = 4
param_combinations = random_search(param_distributions, n_iter=n_iter, seed=83)

results = []
for i, params in enumerate(param_combinations):
    trial = i + 1
    val_acc, result = evaluate_params(params, X_train, y_train_one_hot, layer_sizes, k=5, epochs=100, trial=trial, parallel=True)
    results.append((val_acc, params))
    logging.info(f"Trial {trial}: Validation Accuracy: {val_acc:.4f}")

print("\nTrial Summary:")
print("-" * 75)
print(f"{'Trial':<8} {'Batch Size':<15} {'Learning Rate':<15} {'L2 Lambda':<15} {'Optimizer':<10} {'Val Acc':<10}")
print("-" * 75)
for val_acc, params in sorted(results, reverse=True):
    print(f"{trial:<8} {params['batch_size']:<15} {params['learning_rate']:<15} {params['l2_lambda']:<15} "
          f"{params['optimizer']:<10} {val_acc:.4f}")
```
